```python
import easyocr
import matplotlib.pyplot as plt
import cv2
from google.colab import files
from IPython.display import Image
#導入需要的程式庫
```

```python
uploaded_files = files.upload() #輸入檔案一率為jpg檔
#上傳車牌圖片
file_name = list(uploaded_files.keys())[0]
#獲取文件名
file_path = file_name
#使用文件名建構完整的文件路徑
img = cv2.imread(file_path)
#讀取圖片
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
#顯示圖片
```

```python
ocr = easyocr.Reader(['en'])
resce = ocr.readtext(img, allowlist='ABCDEFGHIJKLMOPQRSPUVWXYZ1234567890-')
print(resce)
#識別及顯示圖像中的文本
```

```python
Anser = []
for i in range(len(resce)):
  resceunpack = resce[i]
  if len(resceunpack[1]) == 6:
    print(resceunpack[1])
  elif len(resceunpack[1]) == 7:
    print(resceunpack[1])
  elif len(resceunpack[1]) == 8:
    print(resceunpack[1])
  elif len(resceunpack[1]) == 9:
    print(resceunpack[1])
  else:
    print("No車牌")
#輸出車牌
```
